```python
import pdfplumber
import os

pdf_folder = "../data/pdf"
raw_folder = "../data/raw"

os.makedirs(raw_folder, exist_ok=True)

for pdf_file in os.listdir(pdf_folder):

    if pdf_file.endswith(".pdf"):

        text = ""

        with pdfplumber.open(
            os.path.join(pdf_folder, pdf_file)
        ) as pdf:

            for page in pdf.pages:

                page_text = page.extract_text()

                if page_text:
                    text += page_text + "\n"

        txt_name = pdf_file.replace(".pdf", ".txt")

        with open(
            os.path.join(raw_folder, txt_name),
            "w",
            encoding="utf-8"
        ) as f:
            f.write(text)
```
